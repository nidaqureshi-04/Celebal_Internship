```python
df= pd.read_csv('Iris.csv')
df.head()
```

# 1️⃣ HISTOGRAM of sepal_length

```python
plt.figure(figsize=(6,4))
plt.hist(df['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2️⃣ LINE GRAPH: Mean sepal_length per species

```python
mean_sepal = df.groupby('species')['sepal_length'].mean()
plt.figure(figsize=(6,4))
plt.plot(mean_sepal.index, mean_sepal.values, marker='o', color='green')
plt.title('Mean Sepal Length per Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

# 3️⃣ PIE CHART: Species distribution

```python
species_counts = df['species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Species Distribution')
plt.show()
```

# 4️⃣ ADVANCED PIE CHART: Proportion of sepal length sum by species

```python
sepal_sum = df.groupby('species')['sepal_length'].sum()
plt.figure(figsize=(6,6))
colors = ['gold', 'lightcoral', 'lightskyblue']
plt.pie(sepal_sum, labels=sepal_sum.index, autopct='%1.1f%%', colors=colors, startangle=90, explode=[0.05]*3)
plt.title('Proportion of Sepal Length Sum by Species')
plt.show()
```

# 5️⃣ BOX & WHISKERS PLOT: Sepal length by species

```python
plt.figure(figsize=(6,4))
df.boxplot(column='sepal_length', by='species', grid=False)
plt.title('Sepal Length by Species')
plt.suptitle('')  # Remove default title
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

📌 Conclusion
The visualizations revealed clear differences in sepal length across Iris species.

Iris-virginica has the longest sepal length on average.

The dataset is balanced across species (33.3% each).

Sepal length distributions show setosa has consistently smaller sepals, while virginica and versicolor have more variation.

These insights highlight how sepal length can help distinguish species effectively.
